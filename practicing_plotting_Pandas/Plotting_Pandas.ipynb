```python
from pandas import DataFrame
import matplotlib.pyplot as plt
```

```python
Data = {'Unemployment_rate': [6.1,5.8,5.7,5.7,5.8,5.6,5.5,5.3,5.2,5.2],
       'Stock_Index_Price': [1500,1520,1525,1523,1515,1540,1545,1560,1555,1565]
       }
df = DataFrame(Data,columns = ['Unemployment_rate', 'Stock_Index_Price'])
print(df)
```

```python
df.plot(x='Unemployment_rate', y='Stock_Index_Price', kind = 'scatter')
plt.show()
```

```python
Data_2 = {'Year': [1920,1930,1940,1950,1960,1970,1980,1990,2000,2010],
         'Unemployment_Rate':[9.8,12,8,7.2,6.9,7,6.5,6.2,5.5,6.3]
         }
df_2 = DataFrame(Data_2,columns=['Year', 'Unemployment_Rate'])
df_2.plot(x='Year', y='Unemployment_Rate', kind='line')
plt.show()
```

```python
Data_3 = {'Country': ['USA', 'Canada', 'Germany', 'UK', 'France'],
       'GDP_Per_Capita': [45000,42000,52000,49000,47000]
       }
df_3 = DataFrame(Data_3,columns=['Country','GDP_Per_Capita'])
df_3.plot(x='Country', y='GDP_Per_Capita', kind='bar')
plt.show()
```
